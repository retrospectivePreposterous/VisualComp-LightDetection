```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import pandas as pd

from PIL import Image, ImageOps 

plt.rcParams['figure.figsize'] = [20, 10]  # Plot size
```

# EXPERIMENTS IN PILLOW WITH AUTOCONTRAST

```python
img = Image.open('img1.jpg')

plt.imshow(img)
plt.show()
```

```python
img_cont = ImageOps.autocontrast(img, cutoff = 5, ignore = 5)
img_cont = np.array(img_cont)
img_cont = cv2.cvtColor(img_cont, cv2.COLOR_BGR2GRAY)

plt.imshow(img_cont,cmap='gnuplot2')
plt.show()
```

```python
# Set threshold value
THRESHOLD = 170

# Thresholding
thres= np.copy(img)

thres[thres > THRESHOLD] = 255
thres[thres < 255] = 0

plt.imshow(thres,cmap='gnuplot2')
plt.show()
```

```python
light_ratio = thres.mean() / 255
print ( "The Light Ratio is: " + str(round((light_ratio*100),2)) + "%")
```

# OPENCV DIGITALISATION AND THRESHOLDING IN GRAYSCALE

```python
img = cv2.imread('img1.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img)
plt.show()
```

```python
# Set threshold value
THRESHOLD = 170

# Thresholding
thres= np.copy(img)

thres[thres > THRESHOLD] = 255
thres[thres < 255] = 0

plt.imshow(thres)
plt.show()
```

```python
light_ratio = thres.mean() / 255
print ( "The Light Ratio is: " + str(round((light_ratio*100),2)) + "%")
```

```python
# Histogram representation

vals = [i.sum()/255 for i in thres]
steps = [i for i in range(len(vals))]

plt.bar(steps,vals)
```
